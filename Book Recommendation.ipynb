```python
num_ratings_df=ratings_with_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
avg_ratings_df=ratings_with_books.groupby('Book-Title')['Book-Rating'].mean().reset_index()
```

```python
avg_ratings_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
```

```python
popular_df=avg_ratings_df.merge(num_ratings_df,on='Book-Title')
```

```python
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

```python
x=ratings_with_books.groupby('User-ID').count()['Book-Rating']>200
```

```python
padhe_likhe_users =x[x].index
```

```python
filtered_rating = ratings_with_books[ratings_with_books['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    ind = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[ind])), key=lambda x: x[1], reverse=True)[1:5]
    data = []
    print(similar_items)
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('2nd Chance')
```
